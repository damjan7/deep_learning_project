```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import math as m
```

```python
df_iso = pd.read_csv('results_Isotropic_Gaussian.csv')
df_stu = pd.read_csv('results_Student_T.csv')
df_lap = pd.read_csv('results_Laplace.csv')
df_gam = pd.read_csv('results_Gaussian_Mixture.csv')
df_nig = pd.read_csv('results_Normal_Inverse_Gamma.csv')
df_gss = pd.read_csv('results_Gaussian_Spike_and_Slab.csv')
df_mlu = pd.read_csv('results_Mixed_Laplace_and_Uniform.csv')
```

```python
df_priors = [df_iso, df_stu, df_lap, df_gam, df_nig, df_gss, df_mlu]
priors = ['iso', 'stu', 'lap', 'gam', 'nig', 'gss', 'mlu']


# Use a for loop to add a new column to each dataframe indicating the prior
for df, prior in zip(df_priors, priors):
    df['Prior'] = prior

# Concatenate the dataframes together
df = pd.concat(df_priors,ignore_index=True)
```

```python
df
```

```python
metrics = ['Test Accuracy', 'Test ECE', 'Test AUROC',
'Test Accuracy', 'Test ECE', 'Test AUROC',
'Test Accuracy', 'Test ECE', 'Test AUROC',
'Test Accuracy', 'Test ECE', 'Test AUROC']

sizes = [3750, 15000, 60000, 120000]

network = 'FCNN'

cols = ['Test Accuracy', 'Test ECE', 'Test AUROC']
rows = sizes
```

```python
sns.set(style='white', palette='rainbow', context = 'paper')
```

```python
fig, axs = plt.subplots(4, 3, figsize=(12, 8))
fig.subplots_adjust(hspace=0.8, wspace=0.4)
fig.dpi = 300
for a in range(4):
    size = sizes[a]
    df_now = df.loc[(df['Sample Size'] == size) & (df['Network'] == network)]
    for i in range (12):
        sns.lineplot(x='Temperature',
         y = metrics[i], 
         data = df_now, 
         hue='Prior', 
         ax=axs[a, (i%3)], 
         legend=False,  
         markers=True,
         style='Prior', 
         dashes=False).set(xscale = 'log', ylabel = '')
        
axs[0,0].set_title(cols[0], weight='bold').set_fontsize('10')
axs[0,1].set_title(cols[1], weight='bold').set_fontsize('10')
axs[0,2].set_title(cols[2], weight='bold').set_fontsize('10')
axs[0,0].set_ylabel(sizes[0], rotation = 0, labelpad = 20)
axs[1,0].set_ylabel(sizes[1], rotation = 0, labelpad = 20)
axs[2,0].set_ylabel(sizes[2], rotation = 0, labelpad = 20)
axs[3,0].set_ylabel(sizes[3], rotation = 0, labelpad = 20)




fig.suptitle('Evaluation Metric')
fig.supylabel('Sample Size', rotation=0, x = 0)
```
